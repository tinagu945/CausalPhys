```python
from models.modules_causal_vel import *
from data.AL_sampler import RandomPytorchSampler
from data.datasets import *
from data.dataset_utils import *
import argparse
from torch.utils.data import DataLoader
from utils.functions import *

parser = argparse.ArgumentParser()
parser.add_argument("-f",help="Input image, directory, or npy.")
args = parser.parse_args()
args.dims = 9
args.edge_types = 2
args.decoder_hidden = 256
args.skip_first = True
args.num_atoms = 8
args.decoder_dropout = 0.0
args.self_loop = True
args.suffix = "causal_vel_delta_grouped_46656"  
args.input_atoms = 6
args.variations = 4
args.train_size = None
args.temp = 0.01
args.hard = True
args.prediction_steps = 19
args.control_constraint = 1.0
args.gt_A=True

        
decoder = MLPDecoder_Causal(args).cuda()

train_data = load_one_graph_data(
        'test_'+args.suffix, size=args.train_size, self_loop=args.self_loop, control=True, control_nodes=args.input_atoms, variations=args.variations)
train_sampler = RandomPytorchSampler(train_data)
# batch_size fixed to be args.variations for grouped dataset
train_data_loader = DataLoader(
    train_data, batch_size=args.variations, shuffle=False, sampler=train_sampler)

off_diag = np.ones([args.num_atoms, args.num_atoms])
rel_rec = np.array(encode_onehot(np.where(off_diag)[0]), dtype=np.float32)
rel_send = np.array(encode_onehot(np.where(off_diag)[1]), dtype=np.float32)
rel_rec = torch.FloatTensor(rel_rec).cuda()
rel_send = torch.FloatTensor(rel_send).cuda()
```

```python
def control_loss(msg_hook, control_node, input_nodes):
    """Similar to the loss in Interpretable Intuitive Physics Model
    Assume the dataset is arranged as input vars first, target vars in the bottom

    Args:
        msg_hook (torch.Tensor): shape (time_steps, batch_size, node**2, information_dim)
        control_node (int): which node to exclude since it's the controlled changing node
        input_nodes (int): total number of nodes that need to be controlled.
    """

    mean = msg_hook.mean(dim=1, keepdim=True)
    num_nodes = int(np.sqrt(msg_hook.size(2)))
    # Var same size as msg_hook
    var = (msg_hook-mean)**2
    loss = 0
    control=0
    # Only constrain on the fixed input nodes
    for i in range(msg_hook.size(2)):
        # import pdb
        # pdb.set_trace()
        if i % num_nodes != control_node[0] and i % num_nodes < input_nodes:
            # print(i)
            loss += var[:, :, i, :].sum()
        elif i % num_nodes == control_node[0] and i % num_nodes < input_nodes:
            control += var[:, :, i, :].sum()
    return loss/msg_hook.size(1), control/msg_hook.size(1)
```

```python
args.hard = True
def load_test(weight_path):
    decoder.load_state_dict(torch.load(weight_path)[0])
    decoder.rel_graph = torch.load(weight_path)[1].cuda()
    losses=[]
    ratio = 0
    for batch_idx, all_data in enumerate(train_data_loader):
        if batch_idx<200:
            data, which_node, edge = all_data[0].cuda(), all_data[1].cuda(), all_data[2].cuda()
            output, logits, msg_hook = decoder(data, rel_rec, rel_send,
                                               args.temp, args.hard, args.prediction_steps, [])
            control_constraint_loss, control = control_loss(
                msg_hook, which_node, args.input_atoms)
            control_constraint_loss *= args.control_constraint
            
            losses.append(control_constraint_loss.item())
            x = control.item()*5/(1+control_constraint_loss.item())
            if which_node[0] == 2 or which_node[0] == 3:
                print(which_node[0], x)
                ratio += x
            elif x!=0:
                print('wrong ', which_node[0], x)
        else:
            break
    print(ratio)   
    return msg_hook, losses
    
```
